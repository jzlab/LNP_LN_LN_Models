```python
fps = 30
size = (960, 540) #(540, 960)

fourcc = cv2.VideoWriter_fourcc(*"mp4v")
out = cv2.VideoWriter(
    "/home/rotation/Desktop/LNP_LN_LN_Models/lnp_naturalmovies/Video_2_downsized.mp4",
    fourcc,
    fps,
    size
)

for i,frame in enumerate(video):
    # Convert float → uint8
    if frame.dtype == np.float32:
        frame = (frame * 255).astype(np.uint8)
        if i == 0:
            print("correcting dtype")


    # Handle grayscale safely
    if len(frame.shape) == 2:
        frame = cv2.cvtColor(frame, cv2.COLOR_GRAY2BGR)
        if i == 0:
            print("handling gray --> color conversion")

    # Resize to match writer
    frame = cv2.resize(frame, size, interpolation=cv2.INTER_AREA)

    out.write(frame)

out.release()
```
